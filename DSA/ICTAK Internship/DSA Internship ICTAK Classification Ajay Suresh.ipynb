```python
# Extract unique loan types
loan_types = set()
for loans in data['Type_of_Loan']:
    if loans != 'No data':
        loans = loans.replace(' and ', ', ')
        loan_types.update([loan.strip() for loan in loans.split(',')])

loan_types = sorted(loan_types)

loan_types = [loan for loan in loan_types if loan] # Sort for consistency

print(f"Unique loan types: {loan_types}")
```

```python
for loan_type in loan_types:
    data[loan_type] = 0
```

```python
for idx, row in data.iterrows():
    if row['Type_of_Loan'] != 'No data':
        loans = row['Type_of_Loan'].replace(' and ', ', ')
        for loan in loans.split(','):
            loan = loan.strip()
            if loan in loan_types:
                data.at[idx, loan] = 1
```

```python
data = data.fillna(0)
```

```python
data.info()
```

```python
loan_columns = ['Auto Loan', 'Credit-Builder Loan', 'Debt Consolidation Loan', 'Home Equity Loan', 
    'Mortgage Loan', 'No Data', 'Not Specified', 'Payday Loan', 'Personal Loan', 'Student Loan']

def combine_loans(row):
    loans = [loan for loan in loan_columns if row[loan] == 1]
    if not loans:
        return 'No Data'
    return ', '.join(loans)

# Apply the function to create the new combined column
data['Combined_Loans'] = data.apply(combine_loans, axis=1)

# Create a mapping of unique combinations to numeric labels
unique_combinations = data['Combined_Loans'].unique()
combination_to_label = {combination: idx for idx, combination in enumerate(unique_combinations)}

# Map the combinations to numeric labels
data['Loan_Label'] = data['Combined_Loans'].map(combination_to_label)
```

```python
data = data.drop(columns=loan_columns + ['Combined_Loans'])
```

### Label Encoding and Fixing Oversampling

```python
shuffled_df = data.sample(frac=1, random_state=42).reset_index(drop=True)
df_majority1 = shuffled_df[shuffled_df['Credit_Score'] == 'Standard']
df_majority2 = shuffled_df[shuffled_df['Credit_Score'] == 'Good']
df_minority = shuffled_df[shuffled_df['Credit_Score'] == 'Poor']
 
# Undersample the majority class
df_majority1_undersampled = df_majority1.sample(n=18000, random_state=42)
df_majority2_undersampled = df_majority2.sample(n=17000, random_state=42)
# Combine the undersampled majority class with the minority class
sdf = pd.concat([df_majority1_undersampled, df_majority2_undersampled, df_minority])
df = sdf.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
sns.set(style = 'whitegrid')
plt.rcParams["figure.figsize"] = (12,6)

fig, ax = plt.subplots()
xi = np.arange(1, 25, step=1)
yi = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, yi, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 25, step=1))
plt.ylabel('Cumulative variance (%)')
plt.title('Number of components needed to explain variance')

plt.axhline(y=0.97, color='r', linestyle='--')
plt.axhline(y=0.9, color='black', linestyle='--')
plt.text(0.5, 1, '97% Variance Cut-Off', color = 'red', fontsize=16)
plt.text(1, 0.85, '90% Variance Cut-Off', color = 'black', fontsize=16)
plt.show()
```
