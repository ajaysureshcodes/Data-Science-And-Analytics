```python
from scipy.stats import ttest_ind, mannwhitneyu, normaltest
import seaborn as sns
import matplotlib.pyplot as plt


# Visualize the distribution of Monthly Income for each Job Role using a boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Job Role', y='Monthly Income', data=data)
plt.title('Monthly Income Distribution by Job Role')
plt.show()

# Perform T-Test for Monthly Income between different Job Roles
job_roles = data['Job Role'].unique()
t_test_results = {}

for role1 in job_roles:
    for role2 in job_roles:
        if role1 != role2:
            group1 = data[data['Job Role'] == role1]['Monthly Income']
            group2 = data[data['Job Role'] == role2]['Monthly Income']
            
            # Check for normality assumption before choosing T-Test or Mann-Whitney U Test
            _, p_normal1 = normaltest(group1)
            _, p_normal2 = normaltest(group2)

            if p_normal1 > 0.05 and p_normal2 > 0.05:  # If both groups are normally distributed
                t_stat, p_value = ttest_ind(group1, group2)
                test_type = 'T-Test'
            else:
                u_stat, p_value = mannwhitneyu(group1, group2)
                test_type = 'Mann-Whitney U Test'
            
            key = f"{role1} vs {role2}"
            t_test_results[key] = {'Test Type': test_type, 'P-Value': p_value}

# Display the results
for comparison, result in t_test_results.items():
    print(f"{comparison} - {result['Test Type']} - P-Value: {result['P-Value']}")

# Display the results with conclusions
for comparison, result in t_test_results.items():
    conclusion = 'Significant' if result['P-Value'] < 0.05 else 'Not Significant'
    print(f"{comparison} - {result['Test Type']} - P-Value: {result['P-Value']} - Conclusion: {conclusion}")
```

```python
import pandas as pd
from scipy.stats import ttest_ind, mannwhitneyu, normaltest
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming data is your DataFrame with 'Monthly Income' and 'Job Role' columns
# data = ...

# Perform T-Test for Monthly Income between different Job Roles
job_roles = data['Job Role'].unique()
t_test_results = []

for role1 in job_roles:
    for role2 in job_roles:
        if role1 != role2:
            group1 = data[data['Job Role'] == role1]['Monthly Income']
            group2 = data[data['Job Role'] == role2]['Monthly Income']
            
            # Check for normality assumption before choosing T-Test or Mann-Whitney U Test
            _, p_normal1 = normaltest(group1)
            _, p_normal2 = normaltest(group2)

            if p_normal1 > 0.05 and p_normal2 > 0.05:  # If both groups are normally distributed
                t_stat, p_value = ttest_ind(group1, group2)
                test_type = 'T-Test'
            else:
                u_stat, p_value = mannwhitneyu(group1, group2)
                test_type = 'Mann-Whitney U Test'
            
            comparison = f"{role1} vs {role2}"
            conclusion = 'Significant' if p_value < 0.05 else 'Not Significant'
            t_test_results.append({'Comparison': comparison, 'Test Type': test_type, 'P-Value': p_value, 'Conclusion': conclusion})

# Create DataFrames for Significant and Not Significant results
significant_df = pd.DataFrame([result for result in t_test_results if result['Conclusion'] == 'Significant'])
not_significant_df = pd.DataFrame([result for result in t_test_results if result['Conclusion'] == 'Not Significant'])

# Display the DataFrames
print("Significant Results:")
print(significant_df)

print("\nNot Significant Results:")
print(not_significant_df)
```
